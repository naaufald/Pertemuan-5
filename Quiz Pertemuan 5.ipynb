# Data yang digunakan adalah harga laptop

#### data mencakup nama perusahaan, nama produk, type, screen size, screen resolution, CPU, CPU Type, CPU Frequency, RAM, SSD, GPU Company, GPU Type, berat, dan harga. data diambil dari kaggle.com

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
dataa = pd.read_csv(r'D:\Matana\Semester 3\eksplorasi dan visualisasi data\laptop_price.csv')
dataa
```

```python
print(dataa.dtypes)
```

#### karena data yang diinginkan hanya yang berupa angka, maka kita akan memfilter data yang bukan angka.

```python
selected_data = dataa[['Inches', 'CPU_Frequency (GHz)', 'RAM (GB)', 'Weight (kg)', 'Price (Euro)']]
print(selected_data)
```

```python
cov_matrix = selected_data.cov()
print(cov_matrix)
```

```python
corre_matrix = selected_data.corr()
print(corre_matrix)
```

```python
sns.heatmap(corre_matrix, annot=True, cmap='coolwarm')
plt.title('Heatmap Matriks Korelasi')
plt.show()
```

```python
sns.heatmap(corre_matrix, annot=True, cmap='Blues')
plt.title('Heatmap Matriks Korelasi')
plt.show()
```

#### semakin gelap warna yang dihasilkan, berarti semakin tinggi korelasi antar variabel tersebut.

```python
sns.pairplot(selected_data)
plt.title('Pair Plot Data')
plt.show()
```
